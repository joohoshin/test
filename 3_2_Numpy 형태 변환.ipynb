딥러닝 모델링 시 입력 데이터의 형태를 변환해줘야 하는 경우가 많으며
이에 활용되는 함수를 알아보자

* 이미지 데이터의 경우 대부분 4개의 차원을 가짐
* 텐서플로우는 (데이터갯수, 가로, 세로, 컬러 채널) 형태로 모델링



```python
import numpy as np
```

가상으로 딥러닝에 사용될 이미지 배열을 만들어보자. 

이미지를 읽어오면 0~255 사이의 숫자로 Red, Green, Blue 컬러 채널이 들어오게 된다. 

```python
arr1 = np.random.randint(0,255, (30,30,3))  # 0~255 사이숫자로, 30 x 30 사이즈 이미지 3가지 컬러채널을 만든다. 
```

```python
arr1.shape
```

```python
arr1
```

numpy 배열은 matplotlib를 통해 그림으로 출력이 가능하다. 

랜덤으로 만든 컬러 이미지이므로 색상이 다양한 그림이 출력된다

```python
import matplotlib.pyplot as plt
plt.imshow(arr1)
```

딥러닝 모델을 만들때 불러온 데이터를 모델의 입력 형태와 일치 시켜줘야 한다. 

이런 경우 아래의 함수들을 활용하여 맞출 수 있다

expand_dims: 차원을 추가할 수 있다

* expand_dims(변수, 추가할 위치) 추가할 위치는 0번부터 시작

```python
# 1차원 추가하기
arr2 = np.expand_dims(arr1,0)  # 몇 번째에 추가할지 지정 가능
```

```python
?np.expand_dims  # 물음표를 사용하면 명령어에 대한 설명이 뜬다
```

```python
arr2.shape  # 맨 앞에 차원이 1개 추가되었다.
```

squeeze: 차원 삭제
* squeeze(삭제할 차원)

```python
# 뒤쪽에 차원을 추가해보자
arr3 = np.expand_dims(arr1,3)  # 몇 번째에 추가할지 지정 가능
arr3.shape  # 맨 뒤쪽에 차원을 추가하였다. 
```

```python
# 추가한 차원을 삭제해보자
arr3.squeeze(3).shape  # 3번째 위치의 차원을 삭제 (첫번째 위치는 0번이다) 
```

reshape: 형태 변경
* reshape(행, 열, ..) 새로운 차원 수 지정하면 변경됨

```python
# 1차원 배열을 만들어보자. 
arr4 = np.random.randint(0,9,10)
arr4
```

```python
arr4 = arr4.reshape(2,5)
arr4
```

```python
arr4.shape  # 배열의 형태가 변경되었다.
```

transpose: 순서 변경
* transpose(새로운 위치 입력)  원래 순서는 (0,1,2,3..)

```python
arr3.shape
```

```python
arr5=arr3.transpose(3,0,1,2)  # 마지막(3번)을 맨 앞으로 옮김
```

```python
arr5.shape  # 맨 앞으로 배열 형태가 변경되었다
```
