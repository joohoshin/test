상수 만들기

constant를 사용하여 상수를 저장할 수 있습니다

int32는 32비트 정수형을 의미합니다

```python
arr = tf.constant([[1,2], [3,4]])
```

```python
arr
```

```python
arr+3
```

```python
arr*3
```

Numpy와 동일하게 배열간 연산도 지원합니다

행렬곱 연산도 지원합니다

```python
arr2 = tf.constant([[3,4], [5,6]])
```

```python
arr+arr2
```

```python
arr*arr2
```

```python
arr@arr2
```

```python
tf.argmax(arr)
```

Indexing도 Numpy와 동일한 방식으로 사용합니다

```python
tf.reshape(arr4, [5,2])  
```

tensor는 계산 속도가 중요하여 같은 타입끼리만 연산이 가능합니다. 

**  정수와 실수형이 연산되지 않습니다.
 
tf.cast()를 통해서 타입을 맞춰주어야 합니다. 

.dtype을 통해 데이터 타입 확인 가능합니다. (numpy도 동일합니다)

```python
n1 =tf.constant(10)  # 정수
print(n1.dtype)   # .dtype을 통해 타입 확인 가능
n2 = tf.constant(5.) # 실수
print(n2.dtype)
```

```python
n1+n2  # 타입이 맞지 않아서 에러가 발생함
```

cast()함수를 통해 형태를 변환할 수 있습니다.
* cast(변수, 새로운 데이터 타입)

```python
n1 = tf.cast(n1, tf.float32)
```

```python
n1+n2
```
