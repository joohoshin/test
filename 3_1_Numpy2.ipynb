### 배열 자동 생성
* random을 통해서 자동으로 생성가능함
* randint(start, end, size = (row size, column size)) 형태로 사용 (end는 포함되지 않음)

```python
import numpy as np
```

```python
arr1 = np.random.randint(0,10, size=(2,3))  # 0~9로 이루어진 2 x 3 배열 만들기
print(arr1)

arr2 = np.random.randint(0,10, size = 10)   # 0~9로 이루어진 10개짜리 1차원 배열 만들기
print(arr2)
```

* rand(row size, column size)는 0~1 사이의 값을 랜덤 선택 (1은 미포함)


```python
arr3 = np.random.rand(5,3)
print(arr3)
```

```python
arr4 = np.random.rand(5,3)
print(arr4)
```

### 주요 통계값
* 리스트와 마찬가지로 min(), count() 등 되며
* 평균 및 표준편차도 쉽게 구할 수 있다 mean(), std() 

```python
print(arr4.sum())
print(arr4.min())
print(arr4.max())
print(arr4.mean()) # 평균
print(arr4.std()) # 표준편차
print(arr4.var()) # 분산
```

### 특정 범위 가져오기
* arr[start:end:step]으로 특정 범위 데이터를 가져올 수 있다
  (end 는 미포함)
* 리스트와 동일함
* 0부터 시작, 첫번째 요소는 0번임

```python
arr5 = np.random.randint(10, size = 10)
print(arr5)
```

```python
arr5[:5] # 0~4번째 값을 가져온다. 
```

```python
arr5[1:3]
```
